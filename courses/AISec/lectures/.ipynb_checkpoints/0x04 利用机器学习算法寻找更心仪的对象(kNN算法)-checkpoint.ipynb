## 准备数据

在将上述特征数据输入到分类器之前，必须将待处理数据的格式改变为分类器可以接受的格式。

我们需要设计一段程序来处理输入格式问题。我们可以设计一个函数：
- 名称：file2matrix；
- 功能：处理输入格式问题；
- 输入：文件名字符串；
- 输出：训练样本矩阵和类标签向量。

### 定义file2matrix函数，将文本记录转换为numpy可接受的数据格式

```python
"""将文本记录转换为numpy可接受的数据格式"""
import numpy as np
import operator
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

def file2matrix(filename):
    with open(filename,'r') as f:
        arrayOfLines = f.readlines()
        #print(arrayOfLines)
        numberOfLines = len(arrayOfLines) # 返回文件的行数
        #print("文件的行数：%s" % numberOfLines)
        returnMat = np.zeros((numberOfLines,3)) # 创建Numpy矩阵,并以0来填充
        #print("创建的矩阵：")
        #print(returnMat)
        
        classLabelVector = []
        index = 0
        for line in arrayOfLines:
            """解析文件数据到列表"""
            line = line.strip()
            listFromLine = line.split('\t')
            returnMat[index,:] = listFromLine[0:3]
            #classLabelVector.append(np.int(listFromLine[-1])) #TODO:问题语句
            '''
            if listFromLine[-1] == 'largedose':
                classLabelVector.append(2)
            elif listFromLine[-1] == 'smalldose':
                classLabelVector.append(1)
            else:
                classLabelVector.append(0)
            '''    
            classLabelVector.append(listFromLine[-1])
            index += 1
        #将字符串转换为离散数值largedoses为1，didntlike为0，smalldoses为2
        classLabelVector = LabelEncoder().fit_transform(classLabelVector)
    return returnMat,classLabelVector   

file2matrix("data\dating\datingTestSet.txt")
```

### 数据预处理

下面的表格内容是提取了四组数据：

|序号|每年获得的航班里程数|每周消费的冰淇淋公升数|玩视频所耗时间百分比样本|分类|
|-|-|-|-|-|
|1|400|0.8|0.5|1|
|2|134000|12|0.9|3|
|3|20000|0|1.1|2|
|4|32000|67|0.1|2|

如果想要计算样本3和样本4之间的距离，可以使用下列公式（欧氏几何距离）：

$d = \sqrt[2]{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2} = \sqrt[2]{(67 - 0)^2 + (32000 - 20000)^2 + (0.1 - 1.1)^2}$

很容易看出，上面公式中数值差值最大的属性对计算结果的影响最大。换句话说，“每年获得的航班里程数”对于计算结果的影响将远远大于表2-3中其他两个特征：“玩视频所耗时间百分比”、“每周消费的冰淇淋公升数”。


$d = \sqrt[2]{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2} = \sqrt[2]{(67 - 0)^2 + (32000 - 20000)^2 + (0.1 - 1.1)^2}$


出现这种现象的原因，是因为上述3个等权重的特征的计数范围有很大差异（量纲不同）。

在处理这种不同取值范围的特征值时，我们通常需要将各属性的数值进行**归一化**，即将取值范围处理为 0到1 或 -1到1之间的数值。

### 归一化（normalization）公式

$newValue = \frac{oldValue - min}{max - min}$

其中min和max分别是数据集中的最小特征值和最大特征值。

虽然改变数值取值范围增加了分类器的复杂度，但是为了得到更为准确的结果，归一化是需要的。

下面的代码是归一化公式的实现：
- 名称：norm；
- 功能：对一组数据进行归一化：$newValue = (oldValue - min)/(max - min)$ ；
- 输入：数据集；
- 输出：归一化的数据集，变化范围，最小值。

### 编写归一化函数norm

```python
import numpy as np
def norm(dataSet):
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    normDataSet = np.zeros(np.shape(dataSet))
    m = dataSet.shape[0]
    normDataSet = dataSet - np.tile(minVals,(m,1)) # np.tile函数是对参数1 进行复制，复制参数2指定的行、列次
    normDataSet = normDataSet / np.tile(ranges,(m,1)) # /表示矩阵中元素除法，不是矩阵/矩阵，矩阵除矩阵使用np.solve(A,B)
    
    return normDataSet, ranges, minVals
```

在上面的函数中，我们将每列的最小值放在变量minVals中，将最大值放在变量maxVals中。

其中dataSet.min(0)中的参数0使得函数可以从列中选取最小值，而不是选取当前行的最小值。

norm函数会计算可能的取值范围，并返回处理后的矩阵。

代码中的tile()函数将变量内容复制成输入矩阵同样大小的矩阵。

注意这是具体特征值相除，而对于某些数值处理软件包，注意“/”可以表示矩阵除法。

但在numpy库中，矩阵除法需要使用函数linalg.solve(matA, matB)。

### KNN 算法原理

k邻近（k-Nearest Neighbor，简称kNN）学习是一种常用的监督学习方法。

#### 工作原理

它的工作原理非常简单：给定测试样本（测试数据），基于某种距离度量找出训练集中与其最靠近的k个训练样本，然后基于这k个“邻居”的信息进行预测。

- 在分类任务中，可以使用“投票法”，即选择这k个样本中出现最多的类别标记，作为预测结果；
  - 或者基于距离远近，进行加权投票，距离越近权重越大。
- 在回归任务中，可以使用“平均法”，即将这k个样本的实值输出标记的平均值作为预测结果；
  - 还可以基于距离远近，进行加权平均，距离越近的样本权重越大。

事实上，kNN的训练过程不明显，它是“懒惰学习”（lazy learning）的著名代表。这类学习技术在训练阶段，仅仅是把样本保存起来，然后计算一下距离值，训练时间开销近似为0；待收到测试样本后再进行处理。而其他算法，往往在训练阶段就对样本进行学习处理的方法，被称为“急切学习”（eager learning）。

下面的示意图中显示了k邻近分类器。其中有两个关键需要把握：

- k是一个重要参数，当k取不同值时，分类结果会有显著不同；
- 另一方面，若采用不同的距离计算法方法，则找出的“近邻”可能有显著的差别，从而也会导致分类结果显著不同。

![knn分类器示意图](images\04\knn分类器示意图.png)

我们如果假设距离计算方法时“恰当”的，即能够恰当的找出k个近邻，那我们对“最近邻分类器”（1NN，即k=1）在二分类问题上的性能做一个简单的讨论。

给定测试样本x，若其最近邻样本为z，则最近邻分类器**出错的概率就是x与z类别标记不同的概率**，即：

$P(err) = 1 - \sum_{c \in y}P(c | x)P(c|z)$   ——式（1）

假设样本独立同分布，且对任意x和任意小正数$\delta$，在x附近$\delta$距离范围内总能找到一个训练样本；

换句话说，对任意测试样本，总能在任意近的范围内找到式（1）中的训练样本z，令$c^* = argmax_{c \in y}P(c | x)$（argmax()函数：求得最大函数值时的自变量的值），表示贝叶斯最优分类器的结果，有：

$P(err) = 1 - \sum_{c \in y}P(c|x)P(c|z)$  
　　　　$\simeq 1 - \sum_{c \in y}P^2(c|x)$  
　　　　$\le 1 - P^2(c^*|x)$  
　　　　$ = (1 +  P(c^*|x))(1 -  P(c^*|x))$  
　　　　$\le 2 \times (1 -  P(c^*|x))$  

这个结论在今后我们讲到贝叶斯最优分类器时，会令人吃惊，即虽然1NN分类器极为简单，但其泛化错误率不超过贝叶斯最优分类器的错误率的2倍。

### KNN 算法步骤

对未知类别属性的数据集内的每个点依次执行以下操作:
1. 计算已知类别数据集中的点与当前点之间的距离；
2. 按照距离递增次序排序；
3. 选取与当前点距离最小的k个点；
4. 确定前k个点所在类别的出现频率；
5. 返回前k个点出现频率最高的类别作为当前点的预测分类。

KNN的基本实现如下面的函数所示：
- 名称：classifier_KNN；
- 功能：实现基本的K邻近算法；
- 输入：
  + inX：用于分类的输入向量（数量与数据集的行数相同）
  + dataSet：训练样本集
  + labels：标签向量为labels
  + k：用于选择最近邻居的数目
- 输出：发生频率最高的分类标签。

### 距离的计算

上例中，样例对中心的距离计算使用欧式距离公式：

$d = \sqrt[2]{\sum_{i=1}^{n}(x_2^i - x_1^i)^2} $

```python
def classifier_KNN(inX, dataSet, labels, k):
    """
    功能：使用KNN算法，根据已有标记的训练数据集，判断新数据的分类
    输入：  inX，待分类数据（以向量形式表示）；
            dataset：训练集样本集
            labels：标签向量
            k：最近邻居的数量
            
    输出：发生频率最高的元素标签
    """
    dataSetSize = dataSet.shape[0] # 样本数量
    # 按矩阵形式，计算输入向量与所有训练样本的距离
    diffMat = np.tile(inX, (dataSetSize,1)) - dataSet # 把输入数据复制为dataSetSize行，1列的矩阵，减去 dataset
    sqDiffMat = diffMat ** 2 #矩阵元素操作，所有元素求平方
    sqDistances = sqDiffMat.sum(axis = 1) # 对每行数据，计算各列数值的和
    distances = sqDistances ** 0.5 # 对各行的求和值求平方根
    sortedDistIndicies = distances.argsort() #argsort函数返回的是数组值从小到大的索引值
    classCount = {}
    
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]] #根据索引值取出该样本的标签
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1 # 标签（类别）出现一次，累加1
    #print(classCount)
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)# 按类型出现次数逆序排序
    #sorted函数:sorted(iterable,key,reverse)其中iterable表示可以迭代的对象,  key是一个函数，用来选取参与比较的元素，reverse则是用来指定排序是倒序还是顺序operator模块提供的itemgetter函数用于获取对象的哪些维的数据，这里的key=operator.itemgetter(1)，指使key成为一个获取类型计数值的方法
    # sorted函数返回一个列表
    #print(sortedClassCount)
    return sortedClassCount[0][0] #排序后的sortedClassCount是一个列表，元素形式为(label,countvalue)的元组
```

### 算法正确性的计算

如何计算算法的正确性呢？

简单地，我们可以使用预测错误的样例数除以总的样例数，即计算错误率来衡量算法的正确性。完美分类器的错误率是0，而错误率为1的模型可以看作是不能作出任何正确预测的分类器。

$E(f;D) = \frac{1}{m} \sum_{i=1}^{m}II(f(x_i) \not = y_i)$

注：上式中 $II(f(x_i \not = y_i)$表示括号内条件成立时，函数II值计为1。

### 编写datingClassTest函数，计算并返回错误率

```python
def datingClassTest():
    hoRatio = 0.1
    datingDataMat , datingLabels = file2matrix('data\dating\datingTestSet.txt')
    normMat, ranges, minVals = norm(datingDataMat)
    m = normMat.shape[0]
    numTestVecs = int(m*hoRatio)
    errorCount = 0.0
    for i in range(numTestVecs):
        classifierResult = classifier_KNN(normMat[i,:], normMat[numTestVecs:m,:],datingLabels[numTestVecs:m],5)
        print("The classfier came back with: %d, the real answer is: %d" % (classifierResult,datingLabels[i]))
        if(classifierResult != datingLabels[i]):
            errorCount += 1.0
    print("The total error rate is: %f" % (errorCount / float(numTestVecs)))
    
datingClassTest()
```

上面计算的结果是错误率为5%，这是一个相当不错的结果。

我们可以改变函数datingclassTest内的变量hoRatio和变量k的值，检测错误率是否随着变量值的变化而增加。

随着分类算法的不同、数据集的不同、程序实现的不同，错误率会有不同的结果，差别有时还非常大。

上面的例子中，错误率仅5%，阿宝完全可以输入未知对象的属性信息，计算某个对象是否值得交往：
 
- 讨厌didntlike=0
- 喜欢smalldoses=2
- 非常喜欢largedoses=1

```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors,datasets
# 设置显示中文
from matplotlib.font_manager import FontProperties 
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)

def knn(ffMiles,iceCream,percentTats):
    n_neighbors = 15

    # 导入一些数据
    data = pd.read_csv("data\dating\datingTestSet.txt",header=None,sep='\t',names=['年飞机里程','周冰淇淋升数','游戏耗时比','心仪程度'])

    #data.describe()
    X = data.iloc[:,:3]
    y = data.iloc[:,-1]
    #生成一个邻居分类器实例，并训练数据（拟合数据）
    clf = neighbors.KNeighborsClassifier(n_neighbors,weights="distance")
    clf.fit(X,y)

    predict = clf.predict([[ffMiles,iceCream,percentTats]])   
     
    return predict[0]

def classifyPerson():
    resultList = {'didntLike':'不喜欢','largeDoses':'很感兴趣','smallDoses':'有点意思'}
    ffMiles = float(input("每年获得的航班里程数？"))
    iceCream = float(input("每周消费的冰淇淋公升数？"))
    percentTats = float(input("玩视频所耗时间百分比？"))  
   
    result = knn(ffMiles,iceCream,percentTats)

    print("您可能喜欢这个人：%s" % resultList[result] )
    
classifyPerson()
```
